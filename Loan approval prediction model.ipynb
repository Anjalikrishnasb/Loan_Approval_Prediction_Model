```python
#fill the null values of numerical data type(here we choose median,bcz median gives the least value)
#if there is outliers,we can use median,otherwise use mean
df[ 'LoanAmount']=df['LoanAmount'].fillna(df[ 'LoanAmount'].median())
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df[ 'Credit_History']=df['Credit_History'].fillna(df[ 'Credit_History'].mean())
```

```python
#fill the null values of categorical data type
df[ 'Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df[ 'Married']=df['Married'].fillna(df[ 'Married'].mode()[0])
df[ 'Dependents']=df['Dependents'].fillna(df[ 'Dependents'].mode()[0])
df[ 'Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
#Drop unnecessary columns
col=['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Total_Income']
df.drop(columns=col,axis=1)
```

```python
#Encoding techniques:label,one hot encoding
```

```python
from sklearn.preprocessing import LabelEncoder
cols=['Gender','Married','Education','Dependents','Self_Employed','Property_Area','Loan_Status']
le=LabelEncoder()
for col in cols:
    df[col]=le.fit_transform(df[col])
```

```python
#creating independent features and dependent features
#Here loan_status is the dependent variable,all other variables are independent.
#split the variables into dependent and independent

x=df.drop('Loan_Status',axis=1)
y=df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Logistic Regression
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred_model=model.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy=accuracy_score(y_test,y_pred_model)
accuracy
```

```python
confusion_matrix=confusion_matrix(y_test,y_pred_model)
confusion_matrix
```

```python
score=cross_val_score(model,x,y,cv=5)
score
```

```python
np.mean(score)*100
```

```python
#Decision tree
model2=DecisionTreeClassifier()
model2.fit(x_train,y_train)
y_pred_model2=model2.predict(x_test)
accuracy=accuracy_score(y_test,y_pred_model2)
accuracy
```

```python
score=cross_val_score(model2,x,y,cv=5)
print("cross value of decision_tree")
np.mean(score)*100
```

```python
#Random forest
model3=RandomForestClassifier()
model3.fit(x_train,y_train)
y_pred_model3=model3.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy=accuracy_score(y_test,y_pred_model3)
accuracy
```

```python
score=cross_val_score(model3,x,y,cv=5)
print("cross value of random forest is")
np.mean(score)*100
```

```python
model4=KNeighborsClassifier(n_neighbors=3)
model4.fit(x_train,y_train)
y_pred=model4.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
score=cross_val_score(model4,x,y,cv=5)
print("cross value of k neighbors")
np.mean(score)*100
```

```python
from sklearn.metrics import classification_report
def generate_classification_report(model_name,y_test,y_pred):
    report=classification_report(y_test,y_pred)
    print(f"classification report for {model_name}:\n{report}\n")
    
generate_classification_report(model,y_test,y_pred_model)
generate_classification_report(model2,y_test,y_pred_model2)
generate_classification_report(model3,y_test,y_pred_model3)
generate_classification_report(model4,y_test,y_pred)
```

```python
#oversampling,it's mainly used to addres the issue of class imbalance in datasets.
from imblearn.over_sampling import RandomOverSampler
```

```python
oversample=RandomOverSampler(random_state=42)
x_resampled,y_resampled=oversample.fit_resample(x,y)
df_resampled=pd.concat([pd.DataFrame(x_resampled,columns=x.columns),pd.Series(y_resampled,name="Loan_Status")],axis=1)
```

```python
x_resampled_train,x_resampled_test,y_resampled_train,y_resampled_test=train_test_split(x_resampled,y_resampled,test_size=0.2,random_state=42)
```

```python
#logistic
model=LogisticRegression()
model.fit(x_resampled_train,y_resampled_train)
y_pred_model=model.predict(x_resampled_test)
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy=accuracy_score(y_resampled_test,y_pred_model)
accuracy
```

```python
#decision tree
model2=DecisionTreeClassifier()
model2.fit(x_resampled_train,y_resampled_train)
y_pred_model2=model2.predict(x_resampled_test)
accuracy=accuracy_score(y_resampled_test,y_pred_model2)
accuracy
```

```python
#Random forest
model3=RandomForestClassifier()
model3.fit(x_resampled_train,y_resampled_train)
y_pred_model3=model3.predict(x_resampled_test)
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy=accuracy_score(y_resampled_test,y_pred_model3)
accuracy
```

```python
#KNeighbors
model4=KNeighborsClassifier(n_neighbors=3)
model4.fit(x_resampled_train,y_resampled_train)
y_pred=model4.predict(x_resampled_test)
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy=accuracy_score(y_resampled_test,y_pred)
accuracy
```
